# Scaling - For positive Scaling Values





```python
s = [[1.75,1.75],[1.5,1],[1,1.75]]
```

```python
h = arr.shape[0]
w = arr.shape[1]
channel = arr.shape[2]
```

```python
h,w,channel
```

Scaled Image

```python
result_pos = []
```

```python
for i in s:
  sx = i[0]
  sy = i[1]
  sc_h = int(h*sy)
  sc_w = int(w*sx)
  sc_arr = np.zeros((sc_h,sc_w,channel),dtype=np.uint8)
  for i in range(sc_h):
    for j in range(sc_w):
      orig_x = int(j/sx)
      orig_y = int(i/sy)
      orig_x = min(orig_x, w - 1)
      orig_y = min(orig_y, h - 1)
      sc_arr[i][j] = arr[orig_y,orig_x]
  result_pos.append(sc_arr)
```

# Scaling - For Negative Scaling Values

Screenshot 2025-08-02 161808.png

So in our code below, this flipping logic ensures correct pixel placement for negative scaling, which means flipping in image space.



```python
s2 = [[-1.75,-1.75],[-1.5,1],[1,-1.5]]
```

```python
result_neg = []

for i in s2:
  sx = (i[0])
  sy = (i[1])
  sc_h = int(h*abs(sy))
  sc_w = int(w*abs(sx))
  sc_arr = np.zeros((sc_h,sc_w,channel),dtype=np.uint8)
  for i in range(sc_h):
    for j in range(sc_w):
      orig_x = int(j/abs(sx))
      orig_y = int(i/abs(sy))
      tar_x = sc_w - 1 - j if sx < 0 else j
      tar_y = sc_h - 1 - i if sy < 0 else i
      sc_arr[tar_y][tar_x] = arr[orig_y,orig_x]
  result_neg.append(sc_arr)
```

# Combined Final Code

```python
sf = [[1.75,1.75],[1.5,1],[1,1.75],[-1.75,-1.75],[-1.5,1],[1,-1.5]]

for a in range(2):
  arr = plt.imread(f"{a+1}.jpg")
  if arr.dtype != np.uint8:
    arr = (arr * 255).astype(np.uint8)
  h = arr.shape[0]
  w = arr.shape[1]
  channel = arr.shape[2]
  print(channel)
  result = []

  for i in sf:
    sx = (i[0])
    sy = (i[1])
    sc_h = int(h*abs(sy))
    sc_w = int(w*abs(sx))
    sc_arr = np.zeros((sc_h,sc_w,channel),dtype=np.uint8)
    for i in range(sc_h):
      for j in range(sc_w):
        orig_x = int(j/abs(sx))
        orig_y = int(i/abs(sy))
        tar_x = sc_w - 1 - j if sx < 0 else j
        tar_y = sc_h - 1 - i if sy < 0 else i
        sc_arr[tar_y][tar_x] = arr[orig_y,orig_x]
    result.append(sc_arr)

    for i in range(len(result)):
      plt.imsave(f"Scaled_image{a+1}({sf[i][0]},{sf[i][1]}).jpg",result[i])
```
