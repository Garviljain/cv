```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
angles = [60, 45, 30, 15, -15, -45, -60] #rotation around origin - shift centre to origin then rotation then again shift back

for a in range(2):
  arr = plt.imread(f"{a+1}.jpg")
  if arr.dtype != np.uint8:
    arr = (arr * 255).astype(np.uint8)
  h = arr.shape[0]
  w = arr.shape[1]
  channel = arr.shape[2]
  cx, cy = w // 2, h // 2
  result = []

  for i in angles:
    rad = math.radians(i)
    cos_theta = math.cos(rad)
    sin_theta = math.sin(rad)

    ro_w = int(abs(h * sin_theta) + abs(w * cos_theta))
    ro_h = int(abs(h * cos_theta) + abs(w * sin_theta))
    ro_arr = np.zeros((ro_h,ro_w,channel),dtype=np.uint8)
    ro_cx, ro_cy = ro_w // 2, ro_h // 2

    for i in range(h):
      for j in range(w):

      # Shift origin to center
        x0 = j - cx
        y0 = i - cy

        new_x = x0 * cos_theta - y0 * sin_theta
        new_y = x0 * sin_theta + y0 * cos_theta

      # Shift back to new image
        new_x = int(new_x + ro_cx)
        new_y = int(new_y + ro_cy)

        if 0 <= new_x < ro_w and 0 <= new_y < ro_h:
          ro_arr[new_y][new_x] = arr[i,j]

    result.append(ro_arr)

  for i in range(len(result)):
    plt.imsave(f"Rotated_image{a+1}({angles[i]}).jpg",result[i])
    print(f"Rotated_image{a+1}({angles[i]}).jpg saved succesfully!!")
```
