```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
for a in range(2):
  arr = plt.imread(f"{a+1}.jpg")
  if arr.dtype != np.uint8:
    arr = (arr * 255).astype(np.uint8)
  h = arr.shape[0]
  w = arr.shape[1]
  channel = arr.shape[2]

  reflected_x = np.zeros_like(arr)
  reflected_y = np.zeros_like(arr)

  for i in range(h):
      for j in range(w):
          reflected_x[h - 1 - i, j] = arr[i, j]

  for i in range(h):
      for j in range(w):
          reflected_y[i, w - 1 - j] = arr[i, j]

  plt.imsave(f"Reflected_X_axis_img{a+1}.jpg", reflected_x)
  print(f"✅ Reflected_X_axis_img{a+1}.jpg saved.")

  plt.imsave(f"Reflected_Y_axis_img{a+1}.jpg", reflected_y)
  print(f"✅ Reflected_Y_axis_img{a+1}.jpg saved.")
```
