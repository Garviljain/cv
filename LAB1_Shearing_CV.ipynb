```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
angles = [60, 45, 30, 15, -15, -45, -60] #shearing horizontally

for a in range(2):
  arr = plt.imread(f"{a+1}.jpg")
  if arr.dtype != np.uint8:
    arr = (arr * 255).astype(np.uint8)
  h = arr.shape[0]
  w = arr.shape[1]
  channel = arr.shape[2]
  result = []

  for i in angles:
    shear = math.tan(math.radians(i))
    sh_h = h
    sh_w = int(w + abs(shear) * h)
    sh_arr = np.zeros((sh_h,sh_w,channel),dtype=np.uint8)
    for i in range(h):
      for j in range(w):
        orig_x = int(j + shear * i)
        orig_y = i
        if shear < 0:
                orig_x += int(abs(shear)*h)
        sh_arr[i][orig_x] = arr[i,j]
    result.append(sh_arr)

  for i in range(len(result)):
    plt.imsave(f"Sheared_image{a+1}({angles[i]}).jpg",result[i])
    print(f"Sheared_image{a+1}({angles[i]}).jpg saved succesfully!!")
```
